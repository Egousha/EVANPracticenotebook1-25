## 8.Create a function called "calculation" to display `sum, min, max, mean, standard deviation, median calculations about a NumPy array using PrettyTable. Test it for array1,2,3


```python
def calculation(array):
    PT=PrettyTable()
    PT.title=["Numpy array calculations"]
    PT.field_names=["sum","min","max","mean","std",'median']
    PT.add_row([array.sum(), array.min(),array.max(),array.mean(),np.std(array),np.median(array)])
    print(PT)
calculation(array1)
```

```python
def arithmatic(a,b):
    PT=PrettyTable()
    PT.title=["Numpy array arithmatic operation"]
    PT.field_names=["subtraction","addition","multiplication"]
    PT.add_row([np.subtract(a,b), np.add(a,b), np.multiply(a,b)])
    print(PT)
arithmatic(array1,array2)
```
